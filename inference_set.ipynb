### Load Fitur, Scaler, Encoder, dan Model

```python
import pickle
import json
from tensorflow.keras.models import load_model
import pandas as pd
import numpy as np
```

```python
with open("scaler.pkl", "rb") as file_1:
    scaler = pickle.load(file_1)

with open("encoder.pkl", "rb") as file_2:
    encoder = pickle.load(file_2)

with open("num_columns.txt", "r") as file_3:
    num_columns = json.load(file_3)

with open("cat_columns.txt", "r") as file_4:
    cat_columns = json.load(file_4)

model_ann = load_model("model_ann.h5")
```

### Data Inference

```python
# Load data inference
data_inf = pd.read_csv("inference_set.csv")
data_inf
```

### Scaling Encoding Data Inference

```python
data_inf_scaled = scaler.transform(data_inf[num_columns])
data_inf_encoded = encoder.transform(data_inf[cat_columns])
```

```python
data_inf_final = np.concatenate([data_inf_scaled, data_inf_encoded], axis=1)
```

### Predict Data Inference

```python
# Prediksi target data inference
y_pred_inf = model_ann.predict(data_inf_final)
y_pred_inf = np.where(y_pred_inf >=0.5, 1, 0)
y_pred_df = pd.DataFrame(y_pred_inf, columns=["churn_pred"])
y_pred_df
```

```python
# Concat prediksi dengan data inference
data_inf_pred_df = pd.concat([y_pred_df, data_inf], axis=1)
data_inf_pred_df
```

Dari hasil prediksi menggunakan data inference, sebagian besar customer diprediksi churn (terindikasi akan berhenti menggunakan produk) dibandingkan dengan yang tidak churn. Dengan jumlah customer yang churn adalah 6 customer dan 4 customer yang tidak churn.
